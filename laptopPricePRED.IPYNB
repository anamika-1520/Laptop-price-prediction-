```python
df['Memory']=df['Memory'].astype(str).replace('\.0','', regex=True)  # Remove '.0' from Memory values
df['Memory']=df['Memory'].str.replace('GB','', regex=True)  # Remove 'GB' from Memory values
new=df['Memory'].str.split('+', n=1, expand=True)  # Split Memory into two parts

df['First']=new[0]# First part of Memory
df['First']=df['First'].str.strip() # Remove leading/trailing spaces
df['Second']=new[1]  # Second part of Memory

df['Layer1HDD']=df['First'].apply(lambda x: 1 if 'HDD' in x else 0)  # Check if 'HDD' is present
df['Layer1SSD']=df['First'].apply(lambda x: 1 if 'SSD' in x else 0)  # Check if 'SSD' is present
df['Layer1Flash_storage']=df['First'].apply(lambda x: 1 if 'Flash' in x else 0)  # Check if 'Flash' is present
df['Layer1Hybrid']=df['First'].apply(lambda x: 1 if 'Hybrid' in x else 0)  # Check if 'Hybrid' is present

df['First']=df['First'].str.replace(r'\D','')  # Remove non-digit characters from First part
  # Remove non-digit characters from Second part

df['Layer2HDD']=df['Second'].apply(lambda x: 1 if isinstance(x,str) and 'HDD' in x else 0)

df['Layer2SSD']=df['Second'].apply(lambda x: 1 if isinstance(x,str) and 'SSD' in x else 0)
df['Layer2Flash_storage']=df['Second'].apply(lambda x: 1 if isinstance(x,str) and 'Flash' in x else 0)
df['Layer2Hybrid']=df['Second'].apply(lambda x: 1 if isinstance(x,str) and 'Hybrid' in x else 0)



df['Second'].str.replace(r'\D','')
df['First'] = df['First'].fillna('').astype(str).str.extract('(\d+)').fillna(0).astype(int)
df['Second'] = df['Second'].fillna('').astype(str).str.extract('(\d+)').fillna(0).astype(int)



df['HDD']=(df['First']*df['Layer1HDD']) + (df['Second']*df['Layer1HDD'])  # Calculate HDD
df['SSD']=(df['First']*df['Layer1SSD']) + (df['Second']*df['Layer1SSD'])  # Calculate SSD
df['Flash_storage']=(df['First']*df['Layer1Flash_storage']) + (df['Second']*df['Layer1Flash_storage'])  # Calculate Flash Storage
df['Hybrid']=(df['First']*df['Layer1Hybrid']) + (df['Second']*df['Layer1Hybrid'])  # Calculate Hybrid Storage
df.drop(columns=['First',"Second","Layer1HDD","Layer2HDD","Layer1SSD","Layer2SSD","Layer1Flash_storage","Layer2Flash_storage","Layer1Hybrid","Layer2Hybrid"],inplace=True)
```
